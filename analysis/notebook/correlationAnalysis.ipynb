```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import norm
from sklearn.preprocessing import StandardScaler
from scipy import stats
import warnings
from IPython.display import display
warnings.filterwarnings('ignore')
%matplotlib inline
sns.set(rc={"figure.figsize": (12, 8)})
```

```python
CSV_PATH = "../../data/csv/"
DF_NAME = "ksou_categorical.csv"
df = pd.read_csv(CSV_PATH + DF_NAME)
df_original = df.copy()
df = df.drop('id', axis = 1)
```

```python
def plot_box(colname, mydf):
    data = pd.concat([mydf[colname], mydf['soldPrice']], axis = 1)
    sns.boxplot(x=colname, y = 'soldPrice', data=data)
    plt.figure()

def scatter_to_price(colname, xlim = None):
    data = pd.concat([df[colname], df['soldPrice']], axis = 1)
    data.plot.scatter(x = colname, y = 'soldPrice', xlim=xlim)
    plt.figure()
```
