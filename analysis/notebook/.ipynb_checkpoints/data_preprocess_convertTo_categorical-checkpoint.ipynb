```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import norm
from sklearn.preprocessing import StandardScaler
from scipy import stats
import warnings
from IPython.display import display
warnings.filterwarnings('ignore')
%matplotlib inline
sns.set(rc={"figure.figsize": (12, 8)})
```

```python
CSV_PATH = "../../data/csv/"
DF_NAME = "ksou_cleaned.csv"
df = pd.read_csv(CSV_PATH + DF_NAME)
df_original = df.copy()
```

```python
def plot_to_price(colname, xlim = None):
    data = pd.concat([df[colname], df['soldPrice']], axis = 1)
    data.plot.scatter(x = colname, y = 'soldPrice', xlim=xlim)
    plt.figure()
# sell price decrease with the increase of distance
plot_to_price('distanceToCBD');
```

Due to the nature of distance to station and distance to CBD, we want to transform these parameters into categorical data

distanceToCBD:
* 0-1
* 1-10
* 10-20
* 20-30
* 30-40
* '>40'


```python
plot_to_price('distanceToStation',)
plot_to_price('distanceToStation', xlim = (0,1.1))
```

From the above figure, we can see several stages:
* 0-1
* 1-4
* 4-7
* 7-10
* '>10'

```python
def plot_box(colname, mydf):
    data = pd.concat([mydf[colname], mydf['soldPrice']], axis = 1)
    sns.boxplot(x=colname, y = 'soldPrice', data=data)
    plt.figure()

def generate_map(granual, maxval):
    result = {(0,1): '0~1'}
    for i in range(1,maxval,granual):
        result[(i,i+granual)] = str(i) + '~' + str(i+granual)
    return result
```

```python

station_map =    {
    (0,1.1): '0-1',
    (1.1,4.1): '1-4',
    (4.1,7.1): '4-7',
    (7.1,10.1): '7-10',
    }

def station_to_categories(value, granual = None, maxval = None):
    global station_map
    if granual:
        station_map = generate_map(granual, maxval)
    for key in station_map:
        if value >= key[0] and value < key[1]:
            return station_map[key]
    return '>' + str(maxval)
```

```python
cbd_map =    {
    (0,1): '0-1',
    (1,10): '1-10',
    (10,20): '10-20',
    (20,30): '20-30',
    (30,40): '30-40',
    }


def cbd_to_categories(value, granual = None, maxval = None):
    global cbd_map
    if granual:
        cbd_map = generate_map(granual, maxval)
    for key in cbd_map:
        if value >= key[0] and value < key[1]:
            return cbd_map[key]
    return '>' + str(maxval)
```

```python
df = df_original.copy()

df['distanceToCBD'] = df['distanceToCBD'].apply(cbd_to_categories, granual =3, maxval =60 )

df['distanceToStation'] = df['distanceToStation'].apply(station_to_categories , granual =2, maxval =20)

plot_box('distanceToCBD',df)
plot_box('distanceToStation',df)
```

```python
df = df_original.copy()
df['distanceToCBD'] = df['distanceToCBD'].apply(cbd_to_categories, granual =10, maxval =60 )

df['distanceToStation'] = df['distanceToStation'].apply(station_to_categories , granual =2, maxval =10)

plot_box('distanceToCBD',df)
plot_box('distanceToStation',df)
```

```python
df = df_original.copy()
df['distanceToCBD'] = df['distanceToCBD'].apply(cbd_to_categories, granual =8, maxval =40 )

df['distanceToStation'] = df['distanceToStation'].apply(station_to_categories , granual =2, maxval =10)

plot_box('distanceToCBD',df)
plot_box('distanceToStation',df)
```

The distanceToStation is probably not very relevant to sales.
However there is a clear indication of distanceToCBD impacts the sale price.

```python
df.to_csv(CSV_PATH+'ksou_categorical.csv', index = False)
```
